```python
#Concatenate the three dataframes:
total_files = pd.concat([file1, file2, file3], ignore_index=True)
display(total_files)
```

```python
#Perform the data cleaning operations mentioned so far in class
#Delete the column education and the number of open complaints from the dataframe.
total_files_modified = total_files.drop(['Education', 'Number of Open Complaints'], axis=1)
display(total_files_modified)
```

```python
#Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type:
def clean_clv(x):
    if (str(x).find('%') != -1):
        return float(str(x).replace('%', ''))*100
    else:
        return x*100

total_files_modified['Customer Lifetime Value'] = total_files_modified['Customer Lifetime Value'].apply(clean_clv)
display(total_files_modified)
```

```python
#Check for duplicate rows in the data and remove if any:
total_files_modified.loc[total_files_modified.duplicated(), :]
```

```python
total_files_modified2 = total_files_modified.drop_duplicates()
display(total_files_modified2)
```

```python
#Filter out the data for customers who have an income of 0 or less:
Total_files_low_income = total_files_modified2[total_files_modified2['Income'] <= 0]
display(Total_files_low_income)
```
